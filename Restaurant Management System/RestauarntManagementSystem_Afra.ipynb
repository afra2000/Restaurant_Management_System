```python
def submit():
    bill.show()
    p1=0
    p2=0
    p3=0
    p4=0
    p5=0
    p6=0
    p7=0
    p8=0
    p9=0
    p10=0
    s1=0
    s2=0
    s3=0
    s4=0
    s5=0
    s6=0
    s7=0
    s8=0
    s9=0
    s10=0
    s=0
    bid=[]
    bill.bill.setText("\t"+"Items"+"\t\t\t"+"Qty"+"\t"+"Rate")
    e="--------------------------------------------------"
    newlabel=format(bill.bill.text()+"\n"+"\t"+e+"\n")
    bill.bill.setText(newlabel)
    if eo.sroll.text() !='':
        p1=int(eo.sroll.text())
        s1=p1*45
        newlabel1=format("\n"+bill.bill.text()+"\n"+"\t"+"Spring Roll"+"\t\t"+str(p1)+"\t"+str(s1))
        bill.bill.setText(newlabel1)
    if eo.roti.text() !='':
        p2=int(eo.roti.text())
        s2=p2*35
        newlabel2=format("\n"+"\t"+bill.bill.text()+"\n"+"\t"+"Roti"+"\t\t\t"+str(p2)+"\t"+str(s2))
        bill.bill.setText(newlabel2)
    if eo.pt.text() !='':
        p3=int(eo.pt.text())
        s3=p3*80
        newlabel3=format("\n"+"\t"+bill.bill.text()+"\n"+"\t"+"Paneer Tikka"+"\t\t"+str(p3)+"\t"+str(s3))
        bill.bill.setText(newlabel3)
    if eo.pbm.text() !='':
        p4=int(eo.pbm.text())
        s4=p4*100
        newlabel4=format("\n"+"\t"+bill.bill.text()+"\n"+"\t"+"Paneer Butter Masala"+"\t"+str(p4)+"\t"+str(s4))
        bill.bill.setText(newlabel4)
    if eo.mnch.text() !='':
        p5=int(eo.mnch.text())
        s5=p5*60
        newlabel5=format("\n"+"\t"+bill.bill.text()+"\n"+"\t"+"Manchurian"+"\t\t"+str(p5)+"\t"+str(s5))
        bill.bill.setText(newlabel5)
    if eo.mg.text() !='':
        p6=int(eo.mg.text())
        s6=p6*60
        newlabel6=format("\n"+"\t"+bill.bill.text()+"\n"+"\t"+"Mushroom grill"+"\t\t"+str(p6)+"\t"+str(s6))
        bill.bill.setText(newlabel6)
    if eo.gb.text() !='':
        p7=int(eo.gb.text())
        s7=p7*60
        newlabel7=format("\n"+"\t"+bill.bill.text()+"\n"+"\t"+"Gobi 65"+"\t\t\t"+str(p7)+"\t"+str(s7))
        bill.bill.setText(newlabel7)
    if eo.bn.text() !='':
        p8=int(eo.roti.text())
        s8=p8*20
        newlabel8=format("\n"+"\t"+bill.bill.text()+"\n"+"\t"+"Butter Naan"+"\t\t"+str(p8)+"\t"+str(s8))
        bill.bill.setText(newlabel8)
    if eo.dks.text() !='':
        p9=int(eo.dks.text())
        s9=p9*35
        newlabel9=format("\n"+"\t"+bill.bill.text()+"\n"+"\t"+"Drinks"+"\t\t\t"+str(p9)+"\t"+str(s9))
        bill.bill.setText(newlabel9)
    if eo.dsrt.text() !='':
        p10=int(eo.dsrt.text())
        s10=p10*40
        newlabel10=format("\n"+"\t"+bill.bill.text()+"\n"+"\t"+"Desserts"+"\t\t\t"+str(p10)+"\t"+str(s10))
        bill.bill.setText(newlabel10)
    s=s1+s2+s3+s4+s5+s6+s7+s8+s9+s10
    name=eo.cname.text()
    bill.cname.setText(name)
    today=date.today().strftime('%d/%m/%Y')
    bill.date.setText(today)
    e="--------------------------------------------------"
    newlabel=format(bill.bill.text()+"\n"+"\t"+e+"\n")
    bill.bill.setText(newlabel)
    op1=format("\n"+bill.bill.text()+"\n"+"\t"+"Total Amount"+"\t\t\t"+str(s))
    bill.bill.setText(op1)
    con.execute("insert into orders (cname,cost) values('{}','{}')".format(name,s))
    con.commit()
    res=con.execute("select * from orders")
    for i in res.fetchall():
        for j in range(0,3):
            bid.append(i[j])
    a=bid[-3]
    bill.ordno.setText(str(a))
def clear():    
    eo.sroll.setText('')
    eo.roti.setText('')
    eo.pt.setText('')
    eo.pbm.setText('')
    eo.mnch.setText('')
    eo.mg.setText('')
    eo.gb.setText('')
    eo.bn.setText('')
    eo.dks.setText('')
    eo.dsrt.setText('')
    eo.cname.setText('')
def addempl():
    emname=emp.ename.text()
    eage=emp.age.text()
    edesg=emp.dsgn.text()
    if emname=='' or eage=='' or edesg=='':
        msg=QMessageBox()
        msg.setWindowTitle("Error")
        msg.setText("Enter all the details")
        msg.setIcon(QMessageBox.Information)
        x=msg.exec_()
    else:
        con.execute("insert into Employee (emname,age,designation) values('{}','{}','{}')".format(emname,eage,edesg))
        con.commit()
        msg=QMessageBox()
        msg.setWindowTitle("Success")
        msg.setText("Added Successfully")
        msg.setIcon(QMessageBox.Information)
        x=msg.exec_()
        emp.ename.setText('')
        emp.age.setText('')
        emp.dsgn.setText('')
def delete():
    eid=emp.did.text()
    flag=0
    if eid=='':
        msg=QMessageBox()
        msg.setWindowTitle("Error")
        msg.setText("Enter ID to be deleted")
        msg.setIcon(QMessageBox.Information)
        x=msg.exec_()
    else:
        a1=con.execute("select * from employee")
        for i in a1.fetchall():
            if(i[0]==int(eid)):
                flag=1
                break
        if flag==1:
            con.execute(("delete from Employee where empno={}").format(eid))
            con.commit()
            msg=QMessageBox()
            msg.setWindowTitle("Success")
            msg.setText("Deleted Successfully")
            msg.setIcon(QMessageBox.Information)
            x=msg.exec_()
        if flag==0:
            msg=QMessageBox()
            msg.setWindowTitle("Error")
            msg.setText("Enter Valid ID")
            msg.setIcon(QMessageBox.Information)
            x=msg.exec_()
        
    emp.did.setText('')
def search1():
    eid=emp.sid.text()
    flag1=0
    if eid=='':
        msg=QMessageBox()
        msg.setWindowTitle("Error")
        msg.setText("Enter ID to be searched")
        msg.setIcon(QMessageBox.Information)
        x=msg.exec_()
    else:
        a1=con.execute("select * from employee")
        for i in a1.fetchall():
            if(i[0]==int(eid)):
                flag1=1
                break
    
        if flag1==1:
            res1=con.execute(("select * from employee where empno={}").format(eid))
            for i in res1.fetchall():
                emp.empdetails.setText("\t"+"Employee ID"+"\t"+"Name"+"\t "+"Age"+"\t"+"Designation")
                e="------------------------------------------------------------------"
                newlabel=format(emp.empdetails.text()+"\n"+"\t"+e+"\n")
                emp.empdetails.setText(newlabel)
                newlabel=format("\n"+emp.empdetails.text()+"\n"+"\t"+str(i[0])+"\t\t"+str(i[1])+"\t"+str(i[2])+"\t"+str(i[3]))
                emp.empdetails.setText(newlabel)  
        if flag1==0:
            msg=QMessageBox()
            msg.setWindowTitle("Error")
            msg.setText("Enter Valid ID")
            msg.setIcon(QMessageBox.Information)
            x=msg.exec_()
    emp.sid.setText('')
def display():
    emp.empdetails.setText("\t"+"Employee ID"+"\t"+"Name"+"\t"+"Age"+"\t"+"Designation")
    e="------------------------------------------------------------------"
    newlabel=format(emp.empdetails.text()+"\n"+"\t"+e+"\n")
    emp.empdetails.setText(newlabel)
    res=con.execute("select * from Employee")
    for i in res.fetchall():
        newlabel=format("\n"+emp.empdetails.text()+"\n"+"\t"+str(i[0])+"\t\t"+str(i[1])+"\t "+str(i[2])+"\t"+str(i[3]))
        emp.empdetails.setText(newlabel)
        #print(i[0],i[1],i[2],i[3])
def clear1():
    emp.empdetails.setText('')
def emp1():
    emp.show()
    emp.addemp.clicked.connect(addempl)
    emp.delete_2.clicked.connect(delete)
    emp.search.clicked.connect(search1)
    emp.disp.clicked.connect(display)
    emp.clearemp.clicked.connect(clear1)
```
